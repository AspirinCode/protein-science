```python
data = {'crystal': [-3.98, -0.66, -3.32], 
        'pose 1': [-0.5056, 1.2293, -1.7349],
        'pose 2': [-1.3110, 1.1621, -2.4731]}
labels = ['Final Intermolecular Energy in kcal/mol',
          'vdW + Hbond + desolv Energy in kcal/mol',
          'Electrostatic Energy in kcal/mol']
```

```python
%matplotlib inline
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import pyplot as plt


fig = plt.figure(figsize=(6,6))
ax = ax=Axes3D(fig)

colors = ['green', 'blue', 'red']
markers = ['s', 'o', '^']

for lab,mar,col in zip(data.keys(), markers, colors):
    ax.scatter(data[lab][1], data[lab][2], data[lab][0], 
           marker=mar, color=col, s=140, label=lab, alpha=0.7)


ax.set_xlabel(labels[1])
ax.set_ylabel(labels[2])
ax.set_zlabel(labels[0])
ax.legend()

plt.title('Intermolecular energy components')
     
plt.show()
```

```python
import matplotlib

fig = plt.figure(figsize=(6,6))

colors = ['green', 'blue', 'red']

lab1 = matplotlib.lines.Line2D([0],[0], linestyle="none", c=colors[0], marker=markers[0])
lab2 = matplotlib.lines.Line2D([0],[0], linestyle="none", c=colors[1], marker=markers[1])
lab3 = matplotlib.lines.Line2D([0],[0], linestyle="none", c=colors[2], marker=markers[2])
plt.legend([lab1,lab2,lab3], data.keys(), numpoints=1, prop={'size':16}, fontsize=16)

plt.plot()
plt.show()
```

```python
labels = ['Estimated Free Energy of Binding',
          'Final Intermolecular Energy',
          'Final Total Internal Energy',
          'Torsional Free Energy',
          "Unbound System's Energy"]

crystal = [-2.19, -3.98, -2.53, 1.79, -2.53] 
pose1 = [1.28, -0.51, -2.55, 1.79, -2.55]
pose2 = [0.48, -1.31, -2.12, 1.79, -2.12]
```

```python
import matplotlib.pyplot as plt


# Setting the positions and width for the bars
pos = list(range(5)) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(8,6))

plt.bar(pos, crystal, width,
                 alpha=0.5,
                 color='r',
                 label=labels[0])

plt.bar([p + width for p in pos], pose1, width,
                 alpha=0.5,
                 color='g',
                 label=labels[1])
    
plt.bar([p + width*2 for p in pos], pose2, width,
                 alpha=0.5,
                 color='b',
                 label=labels[2])

# Setting axis labels and ticks
ax.set_ylabel('kcal/mol')
ax.set_title('Energy components of 3 ligand poses')
ax.set_xticks([p + 1.5 * width for p in pos])
ax.set_xticklabels(labels)

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([min(crystal + pose1 + pose2) * 1.5, max(crystal + pose1 + pose2) * 1.5])

# Adding the legend and showing the plot
plt.legend(['crystal', 'pose 1 (RMSD 3.649 A)', 'pose 2 (RMSD 2.998 A)'], loc='lower right')

# rotate axis labels
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')

plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt


# Setting the positions and width for the bars
pos = list(range(3)) 
width = 0.25 
    
# Plotting the bars
fig, ax = plt.subplots(figsize=(8,6))

plt.bar(pos, crystal, width,
                 alpha=0.5,
                 color='r',
                 label=labels[0])

plt.bar([p + width for p in pos], pose1, width,
                 alpha=0.5,
                 color='g',
                 label=labels[1])
    
plt.bar([p + width*2 for p in pos], pose2, width,
                 alpha=0.5,
                 color='b',
                 label=labels[2])

# Setting axis labels and ticks
ax.set_ylabel('kcal/mol')
ax.set_title('Components of the final internal energy')
ax.set_xticks([p + 1.5 * width for p in pos])
ax.set_xticklabels(labels)

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([min(crystal + pose1 + pose2) * 1.5, max(crystal + pose1 + pose2) * 1.5])

# Adding the legend and showing the plot
plt.legend(['crystal', 'pose 1 (RMSD 3.649 A)', 'pose 2 (RMSD 2.998 A)'], loc='lower right')

# rotate axis labels
plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')

plt.grid()
plt.show()
```
